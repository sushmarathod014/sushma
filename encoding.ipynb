# 3.oneHotEncoding using sklearn

```python
df=pd.read_csv('dataset/employees.csv')
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop='first',sparse=False)
df2=ohe.fit_transform(df[['Team']])
df2
```

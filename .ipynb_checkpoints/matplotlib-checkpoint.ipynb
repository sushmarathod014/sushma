```python
#importing the required packages
import matplotlib.pyplot as plt
import numpy as np
```

```python
#creating two arrays
X=np.array([1,2,3,4,5])
Y=X**2
```

```python
#plotting the values
plt.plot(X,Y)
```

```python
#creating sample data for plotting
x = np.arange(1,16)    #1 â€¦ 15
y = x**2    #square of x values
z = (y/2)+40    #arithmetic operation applied to y values
```

```python
#step1 #Creating an empty canvas or figure, to plot the values and visualize the plot
#creating an empty canvas/figure
fig = plt.figure()
plt.show()
```

```python
#step2 #Setting the axes for the figure.
#creating an empty canvas/figure
fig = plt.figure()
#setting axes as [left, bottom, width, height]
ax = fig.add_axes([0,0,1,1])
```

```python
#step3 #Plotting a line on the axes.
#creating an empty canvas/figure
fig = plt.figure()
#setting axes
ax = fig.add_axes([0,0,1,1])
#plotting the lines
ax.plot(x,y,color="g",label='line1' )
plt.show()
```

```python
#step4 #Plotting multiple lines on the axes.
#creating an empty canvas/figure
fig = plt.figure()
#setting axes
ax = fig.add_axes([0,0,1,1])
#plotting the lines
ax.plot(x,y,color="g",label='line1' )
ax.plot(x,z,color="r",label='line2')
plt.show()
```

```python
#step5 #Setting the title for the plot and the labels for the axes.
#creating an empty canvas/figure
fig = plt.figure()
#setting axes
ax = fig.add_axes([0,0,1,1])
#plotting the lines
ax.plot(x,y,color="g", label='line1' )
ax.plot(x,z,color="r", label='line2')
ax.set_title('Title of the Plot',fontsize=15)    #setting title
ax.set_xlabel('Label for X axis',fontsize=15)    # setting X label
ax.set_ylabel('Label for Y axis',fontsize=15)    # setting Y _label
plt.show()
```

```python
#step6 #Adding legend to the plot.

#creating an empty canvas/figure
fig = plt.figure()
#setting axes
ax = fig.add_axes([0,0,1,1])
#plotting the lines
ax.plot(x,y,color="g", label='line1' )
ax.plot(x,z,color="r", label='line2')
ax.set_title('Title of the Plot',fontsize=15)    #setting title
ax.set_xlabel('Label for X axis',fontsize=15)    # setting X label
ax.set_ylabel('Label for Y axis',fontsize=15)    # setting Y _label
ax.legend()    #adding legend
plt.show()
```

```python
#subplots
#creating sample data for plotting
x = np.arange(1,16)    #1 â€¦ 15
y = x**2    #square of x values
z = (y/2)+40    #arithmetic operation applied to y values
#creating an empty canvas/figure
fig = plt.figure(figsize=(8,5))
#setting axes as [left, bottom, width, height]
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.10,0.42,0.45,0.45])
#plotting the lines
ax1.plot(x,y,color="g", label='line1' )
ax1.set_title('Title for Plot 1',fontsize=15)    #setting title
ax1.set_xlabel('Plot 1 - X axis',fontsize=15)    # setting X label
ax1.set_ylabel('Plot 1 - Y axis',fontsize=15)    # setting Y _label
ax2.plot(x,z,color="r", label='line2')
ax2.set_title('Title for Plot 2',fontsize=15)    #setting title
ax2.set_xlabel('Plot 2 - X axis',fontsize=15)    # setting X label
ax2.set_ylabel('Plot 2 - Y axis',fontsize=15)    # setting Y _label
ax1.legend()    #adding legend
ax2.legend()
plt.show()
```

```python
# plt.subplots(nrows, ncols)
#creating sample data for plotting
x = np.arange(1,16)    #1 â€¦ 15
y = x**2    #square of x values
z = (y/2)+40    #arithmetic operation applied to y values
#initializing two axes for the subplot
fig, (ax1, ax2) = plt.subplots(1, 2)    
ax1.plot(x,y,color="g", label='line1' )
ax1.set_title('Title for Plot 1',fontsize=15)    #setting title
ax1.set_xlabel('Plot 1 - X axis',fontsize=15)    # setting X label
ax1.set_ylabel('Plot 1 - Y axis',fontsize=15)    # setting Y _label
ax2.plot(x,z,color="r", label='line2')
ax2.set_title('Title for Plot 2',fontsize=15)    #setting title
ax2.set_xlabel('Plot 2 - X axis',fontsize=15)    # setting X label
ax2.set_ylabel('Plot 2 - Y axis',fontsize=15)    # setting Y _label
plt.show()
```

```python
#creating sample data for plotting
x = np.arange(1,16)    #1 â€¦ 15
y = x**2    #square of x values
z = (y/2)+40    #arithmetic operation applied to y values
#initializing two axes for the subplot
fig, (ax1, ax2) = plt.subplots(1, 2)    
ax1.plot(x,y,color="g", label='line1' )
ax1.set_title('Title for Plot 1',fontsize=15)    #setting title
ax1.set_xlabel('Plot 1 - X axis',fontsize=15)    # setting X label
ax1.set_ylabel('Plot 1 - Y axis',fontsize=15)    # setting Y _label
ax2.plot(x,z,color="r", label='line2')
ax2.set_title('Title for Plot 2',fontsize=15)    #setting title
ax2.set_xlabel('Plot 2 - X axis',fontsize=15)    # setting X label
ax2.set_ylabel('Plot 2 - Y axis',fontsize=15)    # setting Y _label
fig.tight_layout()
```

```python
#creating sample data for plotting
x = np.arange(1,16)    #1 â€¦ 15
y = x**2    #square of x values
z = (y/2)+40    #arithmetic operation applied to y values
#creating an empty canvas/figure
fig = plt.figure(figsize=(9,9))
#setting axes as [left, bottom, width, height]
ax1 = fig.add_axes([0,0,0.4,0.4])
ax2 = fig.add_axes([0.5,0,0.4,0.4])
ax3 = fig.add_axes([0,0.5,0.4,0.4])
ax4 = fig.add_axes([0.5,0.5,0.4,0.4])
#plotting the lines
ax1.plot(x,y,color="g", label='line1' )
ax1.set_title('Title for Plot 1',fontsize=15)    #setting title
ax1.set_xlabel('Plot 1 - X axis',fontsize=15)    # setting X label
ax1.set_ylabel('Plot 1 - Y axis',fontsize=15)    # setting Y _label
ax1.legend()

ax2.plot(x,z,color="r", label='line2')
ax2.set_title('Title for Plot 2',fontsize=15)    #setting title
ax2.set_xlabel('Plot 2 - X axis',fontsize=15)    # setting X label
ax2.set_ylabel('Plot 2 - Y axis',fontsize=15)    # setting Y _label
ax2.legend()

ax3.plot(x,z,color="y", label='line2')
ax3.set_title('Title for Plot 2',fontsize=15)    #setting title
ax3.set_xlabel('Plot 2 - X axis',fontsize=15)    # setting X label
ax3.set_ylabel('Plot 2 - Y axis',fontsize=15)    # setting Y _label
ax3.legend()

ax4.plot(x,z,color="b", label='line2')
ax4.set_title('Title for Plot 2',fontsize=15)    #setting title
ax4.set_xlabel('Plot 2 - X axis',fontsize=15)    # setting X label
ax4.set_ylabel('Plot 2 - Y axis',fontsize=15)    # setting Y _label
ax4.legend()
plt.show()
```
